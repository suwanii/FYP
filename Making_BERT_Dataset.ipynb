```python
import pandas as pd
import json
import spacy
from spacy.gold import docs_to_json, biluo_tags_from_offsets, spans_from_biluo_tags
```

```python
def load_data(file):
    with open (file, "r") as f:
        data = json.load(f)
    return (data)
```

```python
file = r"/content/drive/MyDrive/FYP/annotatedsmallsub"
data = load_data(file)
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
TRAIN_DATA_ = data
```

```python
#TRAIN_DATA_ = data
docs = []
alldata = {}
for text, annot in TRAIN_DATA_:
    doc = nlp(text)
    tags = biluo_tags_from_offsets(doc, annot['entities'])
    print("tags are,", tags)
    print("------------------")
    entities = spans_from_biluo_tags(doc, tags)
    print("entities are, ", entities)
    print("------------------")
    doc.ents = entities
    docs.append(doc)
```

```python
k = docs_to_json(docs)
```

```python
sent = []
n = []
for i in range(len(k['paragraphs'])):
    information = k['paragraphs'][i]['sentences'][0]['tokens']
    #print(information)
    #print("-------------")
    #print(len(information))
    sent = i
    for j in range(len(information)):
        #print(information[j])
        word = information[j]['orth']
        pos = information[j]['tag']
        tag = information[j]['ner']
        n.append((sent,word,pos,tag))
        #print("\\\\\\\\\\\\\\\\\\\\\\")
    #print("//////")
```

```python
df = pd.DataFrame(n,columns = ['Sentence #','text','pos','tag'])
df
```

```python
df['tag'] = df['tag'].str.replace("L-","I-")
df['tag'] = df['tag'].str.replace("U-","B-")
```

```python
df.to_csv(r"/content/drive/MyDrive/FYP/BERTdataset_BIO.csv")
```
